## **Counter : 리스트 원소의 등장 횟수**

- `from collections import Counter`
- 출력 : 내림차순으로 값이 큰 것부터 나온다.
- 상위요소 반환 : `most_common()`
  - `Counter('aaabbbffc').most_common(1)` : 상위요소 하나 출력
  - `Counter('aaabbbffc').most_common(2)` : 상위요소 두개 출력
- 두 카운터의 연산
  - $+$, $-$ 사용
  - `a.subtract(b)` : a counter에서 b counter 빼기
  - `a&b` : 교집합, `a|b` : 합집합
  - 카운터 된 숫자만큼의 요소 리턴
    - `list(counter.elements())` : ['a', 'a', 'a', 'b', 'b']
  - 카운터 key값 리스트 리턴
    - `list(counter)` : ['a', 'b']

````python
from collections import Counter

counter = Counter(['red', 'blue', 'red', 'green', 'blue', 'blue'])
print(counter['blue'])  # blue 등장 횟수
print(dict(counter))  # 색깔별 등장 횟수 dict로 생성
print(sorted(dict(counter).items(), key=lambda x:x[1], reverse=True))  # 등장 횟수로 내림차순 dictionary 정렬
````

## **Join string**

````python
str1 = "I love "
str2 = "Python."
print(''.join([str1, str2]))
````

````python
strs = ['Life', 'is', 'short,', 'I', 'use', 'Python']
def join_strs(strs):
    result = ''
    for s in strs:
        result += ' ' + s
    return result[1:]

print(join_strs(strs))
````

````python
def join_strs_better(strs):
    return ' '.join(strs)
print(join_strs_better(strs))
````

Using the `join()` method can be **4 times faster** than using + to join the strings in the list.

## **최대 공약수 & 최소공배수**

- 최대 공약수
  - math 라이브러리의 `gcd()` 함수
  - 유클리드 호제법
    > **A % B = R (A > B) 일 때**,</br>
A와 B의 최대공약수 = **B와 R의 최대공약수**
- 최소 공배수
  - 두 수의 합 // 최대공약수 : `a + b // math.gcd (a, b)`

````python
import math

# 최소 공배수(LCM)를 구하는 함수
def lcm(a, b):
    return a * b // math.gcd(a, b)

a = 21
b = 14

print(math.gcd(a, b))  # 최대 공약수(GCD) 계산
print(lcm(a, b))  # 최소 공배수(LCM) 계산
````

## **deque**

데이터를 양방향에서 추가하고 제거할 수 있는 자료구조
- `append`

but, 무작위 접근의 시간 복잡도가 $O(N)$이고, 내부적으로 linked list를 사용하고 있기 때문에 N번째 데이터에 접근하려면 N번 순회가 필요하다.

````python
from collections import deque

queue = deque([1, 2, 3])
queue.append(4)
queue
````

````python
deque([1, 2, 3, 4])
print(queue.popleft())
print(queue.popleft())
queue
````

## **itertools : 순열과 조합**

- `from itertools import permutations, combinations, product, combinations_with_replacement`
- `list(permutations(data, r))` 
  - $_{n}P_r$
  - n개의 데이터 리스트 -> r개의 데이터의 순열(순서O)
- `list(combinations(data, r))` 
  - $_{n}C_r$
  - n개의 데이터 리스트 -> r개의 데이터의 조합(순서X)
- `list(product(data, r))` 
  - $_{n}P_r$ + 중복 허용
  - n개의 데이터 리스트 -> r개의 데이터의 순열(순서X, 중복O)
- `list(combinations_with_replacement(data, r))` 
  - $_{n}C_r$ + 중복 허용
  - n개의 데이터 리스트 -> r개의 데이터의 조합(순서X, 중복O)

## **zip(a, b)**

- 두 리스트를 병합해 대응되는 요소끼리 튜플로 합치는 함수

````python
# List

Number = [1,2,3,4]
Name = ['hong','gil','dong','nim']
Number_Name = list(zip(Number,Name))
print(Number_Name)
````

````python
# Dict

Number = [1,2,3,4]
Name = ['hong','gil','dong','nim']
dic = {}
for i in range(len(Number)) : 
    dic[Number[i]] = Name[i]
print(dic)
````

## **deque**

````python
from collections import deque

data = deque([2, 3, 4])

data.appendleft(1)  # 왼쪽 삽입 : deque([1, 2, 3, 4])
data.append(5)  # 오른쪽 삽입 : deque([1, 2, 3, 4, 5])
data.popleft()  # 왼쪽 삭제 : deque([2, 3, 4, 5])
data.pop()  # 오른쪽 삭제 : deque([2, 3, 4])

data
````

````python
array = [[50, "apple"], [30, "banana"] , [400, "melon"]]

### int 기준 정렬 ###

# .sort()
array.sort(key = lambda x:x[0])
print(array)

# sorted()
print(sorted(array, key = lambda x: x[0]))

### str 기준 정렬 ###

# .sort()
array.sort(key = lambda x:x[1])
print(array)

# sorted()
print(sorted(array, key = lambda x: x[1]))
````

````python
### 성적이 낮은 순서로 학생 출력하기 ###
"파이썬 기본 정렬 라이브러리 활용할 것" 

# N 입력 받기
n = int(input())

# N명의 학생 정보를 입력 받아 리스트에 저장
array = []
for i in range(n):
    input_data = input().split()  # .split() : 공백을 기준으로 문자열을 나누어 줌
    # 이름은 문자열 그대로, 점수는 정수형으로 변환하여 저장
    array.append((input_data[0], int(input_data[1])))

# 키(Key)를 이용하여, 점수를 기준으로 정렬
array = sorted(array, key=lambda x: x[1])

# 정렬이 수행된 결과를 출력
for student in array:
    print(student[0], end=' ')
````

## **dict sort**

````python
# key값 기준 정렬

names = {'Mary':10999, 'Sams':2111, 'Aimy':9778, 'Tom':20245, 'Michale':27115, 'Bob':5887, 'Kelly':7885}

def f1(x):
    return x[0]

res = sorted(names.items())  # key와 value 모두 호출
print(res)

res = sorted(names.items(), key=f1, reverse=True)
print(res)
````

````python
# value 기준 정렬

names = {'Mary':10999, 'Sams':2111, 'Aimy':9778, 'Tom':20245, 'Michale':27115, 'Bob':5887, 'Kelly':7885}

def f2(x):
    return x[1]

res = sorted(names.items(), key=f2)  
print(res)

res = sorted(names.items(), key=(lambda x:x[1]), reverse=True)
print(res)
````

- 카운팅정렬(Counting Sort) : $O(N)$
