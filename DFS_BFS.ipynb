```python
#   n!
def factorial_iterative(n):        
    result = 1
    # 1 n   
    for i in range(1, n + 1):
       result *= i
    return result
```

```python
#   n!
def factorial_recursive(n):        
    if n <= 1: # n 1   1 
        return 1
    # n! = n * (n - 1)!   
    return n * factorial_recursive(n - 1)
```

```python
#    n! (n = 5)
print(' :', factorial_iterative(5))
print(' :', factorial_recursive(5))
```

```python
# DFS  

def dfs(graph, v, visited):
    #    
    visited[v] = True
    print(v, end=' ')
    #       
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)

#  :       (2 )
graph = [
  [],  #   1  0 .
  [2, 3, 8],  # 1  
  [1, 7],  # 2  
  [1, 4, 5],
  [3, 5],
  [3, 4],
  [7],
  [2, 6, 8],
  [1, 7]
] 

#       (1 )
visited = [False] * 9  # index 0    +1 

#  DFS  
dfs(graph, 1, visited)
```

# **BFS (Breadth-First Search)**

image.png

- BFS **`deque`**       $O(N)$  .
-   **DFS  **.
-   DFS BFS     .
-  **       **   

```python
from collections import deque

# BFS  
def bfs(graph, start, visited):
    # (Queue)   deque  
    queue = deque([start])
    #    
    visited[start] = True
    #    
    while queue:
        #     
        v = queue.popleft()
        print(v, end=' ')
        #   ,      
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

#       (2 )
graph = [
  [],
  [2, 3, 8],
  [1, 7],
  [1, 4, 5],
  [3, 5],
  [3, 4],
  [7],
  [2, 6, 8],
  [1, 7]
]

#       (1 )
visited = [False] * 9

#  BFS  
bfs(graph, 1, visited)
```

```python
# N, M     
n, m = map(int, input().split())

# 2     
graph = []
for i in range(n):
    graph.append(list(map(int, input())))

# DFS        
def dfs(x, y):
    #      
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False
    #     
    if graph[x][y] == 0:
        #    
        graph[x][y] = 1
        # , , ,     
        dfs(x - 1, y)
        dfs(x, y - 1)
        dfs(x + 1, y)
        dfs(x, y + 1)
        return True
    return False

#  ()   
result = 0
for i in range(n):
    for j in range(m):
        #   DFS 
        if dfs(i, j) == True:
            result += 1

print(result) #  
```

```python
from collections import deque

# N, M     
n, m = map(int, input().split())
# 2     
graph = []
for i in range(n):
    graph.append(list(map(int, input())))

#      (, , , )
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

# BFS  
def bfs(x, y):
    # (Queue)   deque  
    queue = deque()
    queue.append((x, y))
    #    
    while queue:
        x, y = queue.popleft()
        #   4   
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            #      
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            #   
            if graph[nx][ny] == 0:
                continue
            #        
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1
                queue.append((nx, ny))
    #      
    return graph[n - 1][m - 1]

# BFS   
print(bfs(0, 0))
```
